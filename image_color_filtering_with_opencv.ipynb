```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_bonus = cv2.imread("./bonus.jpg")
img_bonus_rgb = cv2.cvtColor(img_bonus,cv2.COLOR_BGR2RGB)
```

```python
type(img_bonus_rgb)
```

```python
plt.imshow(img_bonus_rgb)
plt.axis("off")
```

```python
# Grntde arlkl olarak, kiinin salarn (koyu siyah pikselleri)
# semek iin aadaki forml yazyoruz. Bu formlle, koyu siyah piksellerin
# yani sfra ok yakn piksellerin grnt zerindeki koordinatlarn karyoruz

koordinat_siyah = (img_bonus_rgb[:,:,0] <= 40) & (img_bonus_rgb[:,:,1] <= 40) & (img_bonus_rgb[:,:,2] <= 40)
```

```python
# Turkuaz renginin kodlar = [0,255,255]  yeille mavinin birleimi
img_bonus_rgb[koordinat_siyah] = [0,255,255]
```

```python
plt.imshow(img_bonus_rgb)
plt.axis("off")
```

```python
# futbolcunun tamamn (tm bedenini) turkuaz renge dntrn

# ilk olarak grntdeki beyaz arka plan haricindeki tm renkleri seiyoruz
# NOT: beyaz renginin RGB kodlar [255,255,255] olduu iin
# <255 diyerek beyaz hari tm renkleri al demi oluyoruz
```

```python
img_bonus_rgb_2 = cv2.cvtColor(img_bonus,cv2.COLOR_BGR2RGB)
```

```python
koordinat_beyaz_haric = (img_bonus_rgb_2[:,:,0] < 255) & (img_bonus_rgb_2[:,:,1] < 255) & (img_bonus_rgb_2[:,:,2] < 255)
```

```python
img_bonus_rgb_2[koordinat_beyaz_haric] = [0,255,255]
```

```python
plt.imshow(img_bonus_rgb_2)
plt.axis("off")
```

```python
# krmz renkli formay mavi rengine dntrn

img_bonus_rgb_3 = cv2.cvtColor(img_bonus,cv2.COLOR_BGR2RGB)

koordinat_kirmizi_forma_2 = (img_bonus_rgb_3[:,:,0]>127) & (img_bonus_rgb_3[:,:,1]<80) & (img_bonus_rgb_3[:,:,2]<80)
```

```python
img_bonus_rgb_3[koordinat_kirmizi_forma_2] = [0,0,255]
```

```python
plt.imshow(img_bonus_rgb_3)
#plt.axis("off")
```
